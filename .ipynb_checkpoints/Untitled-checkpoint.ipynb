```python
# Function to convert the board state to a string to use it as a key in the Q-table dictionary.

def board_to_string(board):
    return ''.join(board.flatten())
print('Here\'s what our board looks as a string:')
board_to_string(board)
```

```python
#function that checks if the game is over
def isGameOver(board):

    #check rows for winning condition
    for row in board:
        if len(set(row)) == 1 and row[0] != '-': #checks if all elems are the same and not '-'
            return True, row[0] #returns winning element
    
    #check columns by flipping the board
    for col in board.T:
        if len(col) == 1 and col[0] != '-':
            return True, col[0]
    
```
